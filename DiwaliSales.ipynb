```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape',index_col=0)
```

```python
df.drop(['Status','unnamed1'],axis='columns',inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['Product_ID'].value_counts()
```

```python
df.groupby('Product_ID')['Amount'].mean()
```

```python
ax=sns.catplot(x='Gender',data=df,hue='Gender',kind='count',legend=['F','M'])
for ax in ax.axes.flat:  # Iterate over the axes in the FacetGrid
 for bars in ax.containers:
    ax.bar_label(bars)
```

```python
g=sns.barplot(x='Gender',y='Amount',data=df,hue='Gender',estimator=np.sum)
for bars in g.containers:
    g.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Age Group')
sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Gender')
```

```python
sales_occ=sns.barplot(x="Occupation",y='Amount',data=df,estimator=np.sum,errorbar=None)
plt.xticks(rotation=90)
```

```python
state_sales=df.groupby('State')['Amount'].sum().reset_index().sort_values(by='Amount',ascending=False)
```

```python

sns.barplot(x="State",y='Amount',data=state_sales,hue='State')
plt.xticks(rotation=90)
```

```python
g=sns.catplot(x='State',y='Amount',data=df, kind='bar',estimator=np.sum,order=state_sales['State'],errorbar=None,\
            col='Gender',hue='Marital_Status')
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(90)
plt.show()
```

```python
procat_sales=df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
```

```python
df.groupby('Product_Category')['Amount'].sum().sort_values()
cat_sales=sns.catplot(x='Product_Category',y='Amount',data=procat_sales, kind='bar',errorbar=None,hue='Product_Category')
plt.xticks(rotation=90)

```
